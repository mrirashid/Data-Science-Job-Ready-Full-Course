```python
import numpy as np

# Problem 1: Numerical Partial Derivatives
def f(x, y):
    """The multivariable function to differentiate."""
    return 3 * (x**2) * y + np.sin(x) * np.exp(y)

def approx_derivative_x(func, x, y, h=1e-5):
    """Approximates the partial derivative with respect to x."""
    # TODO : Implement the difference quotient (f(x+h, y) - f(x, y)) / h
    return (func(x + h, y) - func(x, y)) / h

def approx_derivative_y(func, x, y, h=1e-5):
    """Approximates the partial derivative with respect to y."""
    # TODO : Implement the difference quotient (f(x, y+h) - f(x, y)) / h
    return (func(x, y + h) - func(x, y)) / h

x_val = 2.0
y_val = 1.0

df_dx = approx_derivative_x(f, x_val, y_val)
df_dy = approx_derivative_y(f, x_val, y_val)

print("Problem 1 Results:")
print(f"Numerical df/dx at (2.0, 1.0): {df_dx}")
print(f"Numerical df/dy at (2.0, 1.0): {df_dy}\n")
```

```python
# Problem 2: Constructing the Gradient Vector
# TODO : Pack the partial derivatives into a NumPy array to form the gradient
gradient_vector = np.array([df_dx, df_dy])

print("Problem 2 Results:")
print(f"Gradient Vector nabla_f: {gradient_vector}\n")
```

```python
# Problem 3: The Directional Derivative
# A unit vector representing the direction we want to step
u = np.array([0.6, 0.8])

# TODO : Calculate the directional derivative (dot product of gradient and u)
directional_derivative = np.dot(gradient_vector, u)

print("Problem 3 Results:")
if directional_derivative is not None:
    print(f"Directional Derivative in direction u: {directional_derivative:.4f}")
```
