```python
import numpy as np
# Problem 1: Accelerating Matrix Powers
A = np . array ([[0.8 , 0.3] ,
[0.2 , 0.7]])
k = 10

# Method 1: Direct power
A_k_direct = np . linalg . matrix_power (A , k )

# Method 2: Eigendecomposition (Q * Lambda ^k * Q^ -1)
# TODO : Find eigenvalues and eigenvectors of A
eigenvalues , Q = np . linalg . eigh (A)
# TODO : Create the diagonal matrix Lambda and raise its elements to power k
Lambda_k = np . diag (eigenvalues ** k)

# TODO : Reconstruct A^k
Q_inv = np . linalg . inv (Q)
A_k_eigen = Q @ Lambda_k @ Q_inv

print (" Problem 1 Results :")
print ( f" Direct Power :\n{ A_k_direct }\n")
print ( f" Eigendecomposition Power :\n{ A_k_eigen }\n")
```

```python
# Problem 2: The Core of PCA (1D Projection )
np.random.seed(42)
# Generate a dummy dataset of 100 centered samples in 2D
X = np.random.randn(100, 2) @ np.array([[2, 1], [1, 2]])

# TODO : Compute the covariance matrix of X.
# Hint : Use np.cov . Remember to set rowvar = False since columns are features!
cov_matrix = np.cov(X, rowvar=False)

# TODO : Find eigenvalues and eigenvectors of the covariance matrix
evals, evecs = np.linalg.eig(cov_matrix)

# TODO : Identify the principal eigenvector (highest eigenvalue)
principal_idx = np.argmax(evals)
principal_eigenvector = evecs[:, principal_idx]

# TODO : Project X onto the principal eigenvector using the dot product
X_projected = X @ principal_eigenvector

print("Problem 2 Results:")
print(f"Covariance Matrix:\n{cov_matrix}")
print(f"Principal Eigenvector: {principal_eigenvector}")
if X_projected is not None:
    print(f"Shape of compressed data: {X_projected.shape}\n")
```

```python
# Problem 3: The Symmetry Superpower
S = np.array([[3, 1],
              [1, 3]])

# TODO : Find eigenvectors of S
sym_evals, sym_evecs = np.linalg.eigh(S)

# TODO : Calculate the dot product between the first and second eigenvector
# Hint : Extract columns from sym_evecs, e.g., sym_evecs[:, 0]
orthogonality_check = np.dot(sym_evecs[:, 0], sym_evecs[:, 1])

print("Problem 3 Results:")
print(f"Dot product of eigenvectors: {orthogonality_check}")
print("If the result is ~0, they are perfectly orthogonal.")
```
